#Grupo1

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr

sns.set(style="whitegrid")
```

```python
# GRUPO 1 – Carga y exploración de datos
df1 = pd.read_csv("grupo1.csv")

print("Primeras filas del Grupo 1:")
print(df1.head())

print("\n Descripción estadística Grupo 1:")
print(df1.describe())
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(df1['Calificacion'], bins=10, kde=True, color='skyblue')
plt.title('Histograma de Calificaciones - Grupo 1')
plt.xlabel('Calificación Final')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# GRUPO 1 – Regresión y correlación
X1 = df1[['HorasEstudio']]
y1 = df1['Calificacion']

modelo1 = LinearRegression()
modelo1.fit(X1, y1)

pendiente1 = modelo1.coef_[0]
intercepto1 = modelo1.intercept_
r1, _ = pearsonr(df1['HorasEstudio'], df1['Calificacion'])

print(f" Ecuación de regresión Grupo 1: y = {pendiente1:.4f}x + {intercepto1:.4f}")
print(f"Coeficiente de correlación de Pearson: {r1:.4f}")

y_pred1 = modelo1.predict(X1)

plt.figure(figsize=(6, 4))
sns.scatterplot(x='HorasEstudio', y='Calificacion', data=df1, label='Datos')
plt.plot(df1['HorasEstudio'], y_pred1, color='red', label='Línea de regresión')
plt.title('Regresión Lineal - Grupo 1')
plt.xlabel('HorasEstudio')
plt.ylabel('Calificacion')
plt.legend()
plt.show()
```
